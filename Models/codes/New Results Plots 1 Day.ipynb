```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

MUD Share Cluster - Single Day

```python
Share1d = pd.read_excel('New Results Datasets For Plots/MUD Share Cluster - Single Day.xlsx')
Share1d
```

```python
Share1d=Share1d.set_index('Census Division')
Share1d=Share1d.drop('Unnamed: 0',axis=1)
```

```python
df=Share1d.stack().to_frame().reset_index()
df['cluster']=df['level_1'].agg(lambda x: x[8])
df['case']=df['level_1'].agg(lambda x: x[11:-1])
df[0]=df[0]*100
```

```python
fig=plt.figure(figsize=(10,8),dpi=600)
sns.boxplot(y=0,x='cluster',hue='case',data=df,palette="Set3")
plt.xticks(fontsize=24)
xlabels = ['home-work', 'home-bound', 'outgoing']
plt.xticks(range(3), xlabels)
plt.yticks(fontsize=24)
plt.ylabel('Share(%)',fontsize=24)
plt.xlabel('Clusters',fontsize=24)
plt.grid('True')
plt.legend(fontsize=18,loc='center left',bbox_to_anchor=(1,0.5))
# plt.legend(fontsize=18,loc='best')
plt.title('Expected share of MUD deployment of one-day duration',fontsize=24)
plt.show()
# plt.savefig('newFigures/shareCluster1d_New_Re.png',dpi=600);
```
